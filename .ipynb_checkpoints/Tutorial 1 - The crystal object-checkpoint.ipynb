# Tutorial 1: The Crystal object

In this tutorial, we show how to prepare a crystal object and inspect some of it's properties.

```python
%load_ext autoreload
%autoreload 2
```

```python
import numpy as np
import matplotlib.pyplot as plt

from custom_poling.core.crystal import Crystal
```

### Define the crystal

```python
# Create a crystal object

domain_width = 1
number_domains = 100

crystal = Crystal(domain_width,number_domains,z0=0)
```

```python
# Print some of the crystal's properties

print(f'Crystal length: {crystal.length}')
print(f'Domain wall positions: {crystal.domain_walls}')
print(f'Domain wall middles: {crystal.domain_middles}')
```

### Example 1: Unpoled crystal

```python
# Specify and plot the domain configuration

domain_config_1 = [1]*number_domains

crystal.plot_domains(domain_config_1)
```

```python
# We expect the peak to be around k0=0

k0 = 0

# Compute and plot the phase-matching function (PMF) for an unpoled crystal

k_array_1 = np.arange(k0-5e-1,k0+5e-1,1e-3)

crystal.compute_pmf(domain_config_1,k_array_1)
crystal.plot_pmf()
```

### Example 2: Periodically-poled crystal

```python
# We expect the peak to be around the k0 for periodic poling

k0=np.pi/domain_width

# Specify and plot the domain configuration

domain_config_2 = [1,-1]*int(number_domains/2)
crystal.plot_domains(domain_config_2,n_max=10)

# Compute and plot the phase-matching function (PMF) for an periodically-poled crystal

k_array_2 = np.arange(k0-0.5,k0+0.5,0.001)
crystal.compute_pmf(domain_config_2,k_array_2)
crystal.plot_pmf()
```
